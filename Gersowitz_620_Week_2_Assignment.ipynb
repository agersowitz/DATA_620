When software developers are starting to work in a new environment, they are encouraged to start with a simple “beachhead” program.
In this assignment, you’re asked to create the nodes and edges for a basic graph, such as the Krackhardt kite shown below. (You’re welcome to substitute data of your own choosing).

Below is a quick example of how to produce the krackhardt kite graph in networkx.



```python
import networkx as NetX
from matplotlib import pyplot as plt

Graph = NetX.krackhardt_kite_graph()
NetX.draw_networkx(Graph)
plt.show()
```

To further the example I will build a graph of the hierarchy of my team at work with each person and who they report to. Linda is our manager who is above andrea our supervisor who I report to along with Jen and Adriana and the 3 of us all have 2 staff.

```python
G=NetX.Graph()
G.add_nodes_from(['Linda','Andrea', 'Adam','Adriana','Jen','Kelsey','Cullen','Luke','Jenna','Ashley','Zak'])

G.nodes['Linda']['position'] = 'Manager'
G.nodes['Andrea']['position'] = 'Supervisor'
G.nodes['Adam']['position'] = 'AIC'
G.nodes['Adriana']['position'] = 'AIC'
G.nodes['Jen']['position'] = 'AIC'
G.nodes['Kelsey']['position'] = 'Staff'
G.nodes['Jenna']['position'] = 'Staff'
G.nodes['Cullen']['position'] = 'Staff'
G.nodes['Luke']['position'] = 'Staff'
G.nodes['Ashley']['position'] = 'Staff'
G.nodes['Zak']['position'] = 'Staff'

G.add_edges_from([('Andrea', 'Adam'), ('Andrea', 'Adriana'),('Andrea', 'Jen'),('Adam', 'Cullen'),('Adam', 'Zak'),('Linda', 'Andrea'),('Jen', 'Kelsey'),
                  ('Jen', 'Ashley'),('Adriana', 'Luke'),('Adriana', 'Jenna')])

G['Linda']['Andrea']['weight'] = .2


NetX.draw_networkx(G)
plt.show()
```
